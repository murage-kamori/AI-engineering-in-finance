```python
#Track buys, sells, and costs
for transaction in transactions:
    symbol = transaction["symbol"]
    qty = transaction["quantity"]
    price = transaction["price"]
    trans_type = transaction["type"]
    
    if symbol not in portfolio:
        continue 
    
    if "total_buy_qty" not in portfolio[symbol]:
        portfolio[symbol]["total_buy_qty"] = 0
        portfolio[symbol]["total_sell_qty"] = 0
        portfolio[symbol]["total_cost"] = 0
    
    if trans_type == "buy":
        portfolio[symbol]["total_buy_qty"] += qty
        portfolio[symbol]["total_cost"] += qty * price  
    elif trans_type == "sell":
        portfolio[symbol]["total_sell_qty"] += qty

# compute net quantity for each symbol
for symbol in portfolio:
    total_buy = portfolio[symbol]["total_buy_qty"]
    total_sell = portfolio[symbol]["total_sell_qty"]
    portfolio[symbol]["quantity"] = total_buy - total_sell

print("Portfolio after adding quantity:")
print(portfolio)

```

```python
# Define current prices and compute PNL 
current_prices = {"BTC": 35000, "ETH": 2400}
for symbol in portfolio:
    net_qty = portfolio[symbol]["quantity"]
    current_price = current_prices[symbol]
    # Temporary avg (we'll replace with real avg in Step 4)
    avg_price = portfolio[symbol]["total_cost"] / portfolio[symbol]["total_buy_qty"]
    pnl = net_qty * (current_price - avg_price)
    portfolio[symbol]["pnl"] = round(pnl, 2)  
print("Portfolio after adding PNL (with temp avg):")
print(portfolio)
```

```python
# Add average price 
for symbol in portfolio:
    total_buy_qty = portfolio[symbol]["total_buy_qty"]
    if total_buy_qty > 0:
        avg_price = portfolio[symbol]["total_cost"] / total_buy_qty
        portfolio[symbol]["avg_price"] = round(avg_price, 2)
    else:
        portfolio[symbol]["avg_price"] = 0  
        
# Recompute PNL with the real avg_price for accuracy
for symbol in portfolio:
    net_qty = portfolio[symbol]["quantity"]
    avg_price = portfolio[symbol]["avg_price"]
    current_price = current_prices[symbol]
    pnl = net_qty * (current_price - avg_price)
    portfolio[symbol]["pnl"] = round(pnl, 2)

print("Final portfolio after adding avg_price and PNL:")
print(portfolio)
```

```python
# Import datetime and compute post-threshold buy quantities

from datetime import datetime

threshold_date_str = "2025-06-02"
threshold_date = datetime.strptime(threshold_date_str, "%Y-%m-%d").date()

# Initialize for each symbol
for symbol in portfolio:
    portfolio[symbol]["post_2025_06_02_buy_qty"] = 0

# Loop through transactions
for transaction in transactions:
    symbol = transaction["symbol"]
    trans_date_str = transaction["date"]
    trans_date = datetime.strptime(trans_date_str, "%Y-%m-%d").date()
    qty = transaction["quantity"]
    trans_type = transaction["type"]
    
    if trans_type == "buy" and trans_date > threshold_date:
        portfolio[symbol]["post_2022_06_02_buy_qty"] += qty

# Display for each crypto
print("\nBonus: Quantity purchased after 2022-06-02:")
for symbol in portfolio:
    qty = portfolio[symbol]["post_2025_06_02_buy_qty"]
    print(f"{symbol}: {qty}")
```
