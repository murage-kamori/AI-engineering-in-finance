```python
#access matplotlib and give shortcut
import matplotlib.pyplot as plt
```

```python
#line plot
x=[1,2,3,4,5]
y=[10,20,25,30,40]

plt.plot(x,y)
plt.title('Sales over time')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
 
    #plt.plot(x,y)- drawa a line connecting points(x,y)
    #plt.title()-add title to chart
    #plt.xlabel()-add label to x axis
    #plt.ylabel()-add label to y axis   
    #plt.show()-display the plot-always call last
```

```python
#bar chart
months=['Jan','Feb','Mar','Apr','May']
sales=[150,200,250,300,350]

plt.bar(months,sales,color='orange')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()

#plt.bar creates vertical bars
```

```python
#scatter plot
x= [1,2,3,4,5]
y= [10,20,25,30,40]

plt.scatter(x,y,color='red')
plt.title("Customer Growth")
plt.xlabel("Quarter")
plt.ylabel("New Customers")
plt.show()

 #plt.scatter()-creates individual points without connecting lines
```

```python
#histogram
import numpy as np
data=np.random.normal(1000) #generate 1000 random numbers from normal distribution

plt.hist(data,bins=30,color='green',edgecolor='black')
plt.title('Distribution of Returns')
plt.xlabel('Return')
plt.ylabel('Frequency')
plt.show()

#plt.hist()-groups data into bins and plots frequency of each bin
#bins- number of bars in histogram
```

```python
#customize
plt.plot(x,y, color='green', linestyle='--', marker='o', marker='0')
  #color-line color
    #linestyle-dashed line
    #marker-adds symbols to each data point
```
